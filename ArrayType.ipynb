**ArrayType is a data type in the pyspark.sql.types module used to define arrays (lists) of elements in a DataFrame schema.**

syntax :- ArrayType(elementType, containsNull=True)

```
from pyspark.sql.types import ArrayType, StringType, IntegerType
```











````python
from pyspark.sql.types import StructType, StructField, ArrayType, StringType
from pyspark.sql import SparkSession
from IPython.display import display

spark = SparkSession.builder.appName("ArrayTypeExample").getOrCreate()

data = [
    ("Amrit Singh",["Python", "SQL", "Tableau", "PySpark", "Excel"],["Data Engineering", "Data Analytics", "Data Viz"],"Open_to_Work"),
    ("Neeraj Patel",["React.js","MongoDB", "TypeScript","Node.js"], ["Full-Stack Developer"], "Intern"),
    ("Prakhar Kushwaha", ["HTML", "CSS", "JavaScript"], ["Tech-Lead"], "Full-Time_Hired"),
    ("Shivam Verma", ["MERN Stack Development","ML engineer", "Senior Data Scientist", "Devops Engineer", "Open Source grandmaster"],["CEO-SEXSUX"], "CEO")
]
schema = StructType([
    StructField("name", StringType(), True),
    StructField("skills", ArrayType(StringType()), True),
    StructField("Workprofile", ArrayType(StringType()), True),
    StructField("status", StringType(), True)
])

df = spark.createDataFrame(data, schema)
display(df)

from IPython.display import display
pdf = df.toPandas()
display(pdf)
````

## Note that array() function is different from the ArrayType()

````python
from pyspark.sql.functions import col, array, lit, array_union

df = df.withColumn("skills", array_union(col("skills"), array(lit("SQL"))))
display(df.toPandas())
````
