```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("PySpark Example").getOrCreate()
print(spark)
```

```python
df = spark.read.csv("/content/sample_data/california_housing_test.csv", header=True)
display(df)
df.printSchema()
df.show()
```

```python
df2 = spark.read.format("json").load("/content/sample_data/anscombe.json", header=True)
df2.show()
df2.printSchema()
```

## For Uploading Multiple CSV Files

```python
from pyspark.sql.functions import lit

list_of_files = ["/content/sample_data/mnist_test.csv", "/content/sample_data/mnist_train_small.csv", "/content/sample_data/california_housing_train.csv"]

df3 = spark.read.csv(list_of_files, header=True)
df3 = df3.withColumn("file_name", lit(list_of_files))

df3.show()
df3.printSchema()
```

## Uploading the Folder containing Multiple CSV files with same schema

```python
df4 = spark.read.format("csv").option("header", "true").load(path="/content/sample_data")
df4.show()
df4.printSchema()
```
